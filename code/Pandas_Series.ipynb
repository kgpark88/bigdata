# Pandas 시리즈 (Series) 데이터 구조 
1차원의 배열의 값(values)과 각 값에 대응하는 인덱스(index)를 부여할 수 있는 데이터 구조

```python
import numpy as np
import pandas as pd
```

### List, NumPy Array, Dictionary를 Series로 변환 가능

```python
menu = ['피자', '치킨', '맥주']
price = [20000, 18000, 5000]
arr = np.array([20000, 18000, 5000])
d = {'피자':20000,'치킨':18000,'맥주':5000}
```

### List 사용

```python
pd.Series(data=price)
```

```python
pd.Series(data=price,index=menu)
```

series.png

### NumPy Array 사용

```python
pd.Series(arr)
```

### Dictionary 사용

```python
pd.Series(d)
```
